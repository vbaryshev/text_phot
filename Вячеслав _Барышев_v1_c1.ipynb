### –®–∞–≥ 1. –ó–∞–≥—Ä—É–∑–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ –∏ –ø—Ä–æ–≤–µ–¥–∏—Ç–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö

***–ß—Ç–µ–Ω–∏–µ –¥–∞–Ω—ã—Ö –∏–∑ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏***

```python
def read_df(path: str) -> tuple:
    '''–ß–∏—Ç–∞–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—ã –∏–∑ —É–∫–∞–∑–∞–Ω–Ω–æ–π –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏'''
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è —Ñ–∞–π–ª–æ–≤ –ø–µ—Ä–µ–¥ —á—Ç–µ–Ω–∏–µ–º
    path = Path(path)
    train_path = path / 'train_dataset.csv'
    crowd_path = path / 'CrowdAnnotations.tsv'
    expert_path = path / 'ExpertAnnotations.tsv'
    queries_path = path / 'test_queries.csv'
    images_path = path / 'test_images.csv'
    
    if not train_path.exists():
        raise FileNotFoundError(f"–§–∞–π–ª {train_path} –Ω–µ –Ω–∞–π–¥–µ–Ω.")
    if not crowd_path.exists():
        raise FileNotFoundError(f"–§–∞–π–ª {crowd_path} –Ω–µ –Ω–∞–π–¥–µ–Ω.")
    if not expert_path.exists():
        raise FileNotFoundError(f"–§–∞–π–ª {expert_path} –Ω–µ –Ω–∞–π–¥–µ–Ω.")
    if not queries_path.exists():
        raise FileNotFoundError(f"–§–∞–π–ª {queries_path} –Ω–µ –Ω–∞–π–¥–µ–Ω.")
    if not images_path.exists():
        raise FileNotFoundError(f"–§–∞–π–ª {images_path} –Ω–µ –Ω–∞–π–¥–µ–Ω.")
    
    # –ß—Ç–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö
    df_train = pd.read_csv(train_path)
    df_crowd = pd.read_csv(crowd_path, sep='\t', names=['image', 'query_id', 'share_pos', 'count_pos', 'count_neg'])
    df_expert = pd.read_csv(expert_path, sep='\t', names=['image', 'query_id', 'first', 'second', 'third'])
    df_queries = pd.read_csv(queries_path, index_col=[0], sep='|')
    df_images = pd.read_csv(images_path, sep='|')
    
    return df_train, df_crowd, df_expert, df_queries, df_images
```

***–û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –∏–∑ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–≥–æ –∏ —Ç–µ—Å—Ç–æ–≤–æ–≥–æ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã***

```python
def display_images(path: str, df_train: pd.DataFrame, df_queries: pd.DataFrame):
    '''–û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –∏–∑ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–≥–æ –∏ —Ç–µ—Å—Ç–æ–≤–æ–≥–æ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö'''
    
    samples_train = list(df_train['image'].sample(8))
    samples_test = list(df_queries['image'].sample(8))
    
    fig, axes = plt.subplots(4, 4, figsize=(10, 10))
    fig.tight_layout()
    
    for i in range(8):
        train_image_path = Path(path) / 'train_images' / samples_train[i]
        test_image_path = Path(path) / 'test_images' / samples_test[i]
        
        # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
        ax = axes[i // 4, i % 4]
        image = Image.open(train_image_path)
        ax.imshow(image)
        ax.set_xticks([])
        ax.set_yticks([])
        
        # –û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Ç–µ—Å—Ç–æ–≤—ã—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
        ax = axes[(i // 4) + 2, i % 4]
        image = Image.open(test_image_path)
        ax.imshow(image)
        ax.set_xticks([])
        ax.set_yticks([])
```

```python
# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏
path_to_data = 'C:\\datasets\\image_search\\to_upload'
df_train, df_crowd, df_expert, df_queries, df_images = read_df(path_to_data)
```

```python
def filter_child_images(df_train: pd.DataFrame, restricted_words: list, column_name: str) -> pd.DataFrame:
    '''–ò—Å–∫–ª—é—á–µ–Ω–∏–µ –∑–∞–ø—Ä–µ—â–µ–Ω–Ω—ã—Ö —Å–ª–æ–≤ –∏–∑ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö.'''
    # –°–æ–∑–¥–∞–µ–º –ø–∞—Ç—Ç–µ—Ä–Ω –¥–ª—è –ø–æ–∏—Å–∫–∞ —Å–ª–æ–≤
    pattern = '|'.join(restricted_words)
    
    # –û–ø—Ä–µ–¥–µ–ª—è–µ–º –ø–ª–æ—Ö–∏–µ –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–∏
    bad_comments = df_train[df_train[column_name].str.contains(pattern, case=False, na=False)]
    
    # –ò–∑–≤–ª–µ–∫–∞–µ–º query_id –ø–ª–æ—Ö–∏—Ö –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤ –∏ —É–¥–∞–ª—è–µ–º –ø–æ—Å–ª–µ–¥–Ω–∏–µ –¥–≤–∞ —Å–∏–º–≤–æ–ª–∞
    bad_image_ids = bad_comments['query_id'].str[:-2]
    
    # –§–∏–ª—å—Ç—Ä—É–µ–º –¥–∞—Ç–∞—Å–µ—Ç, –∏—Å–∫–ª—é—á–∞—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è —Å –ø–ª–æ—Ö–∏–º–∏ –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞–º–∏
    filtered_df = df_train[~df_train['image'].isin(bad_image_ids)]
    
    return filtered_df

# –ü—Ä–∏–º–µ–Ω—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é –∫ df_train
df_train_filtered  = filter_child_images(df_train, restricted_words, 'query_text')

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
print(df_train_filtered)
```

```python
# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö –≤ TF-IDF –≤–µ–∫—Ç–æ—Ä—ã
vectorizer = TfidfVectorizer(max_features=5000)
text_vectors = vectorizer.fit_transform(df_train_filtered['query_text'])
```

```python
# –ó–∞–≥—Ä—É–∑–∫–∞ –ø—Ä–µ–¥–æ–±—É—á–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ ResNet50 –∏ —É–¥–∞–ª–µ–Ω–∏–µ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —Å–ª–æ—è
model = resnet50(pretrained=True)
model = torch.nn.Sequential(*list(model.children())[:-1])
```

```python
# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –¥–ª—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
```

```python
def preprocess_image(img_path):
    img = Image.open(img_path).convert('RGB')
    img_tensor = preprocess(img)
    img_tensor = img_tensor.unsqueeze(0)
    return img_tensor
```

```python
def get_image_vector(model, img_path):
    img_tensor = preprocess_image(img_path)
    with torch.no_grad():
        img_vector = model(img_tensor)
    return img_vector.flatten().numpy()
```

```python
# –ü–æ–ª—É—á–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–æ–≤ –¥–ª—è –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –∏–∑ –æ–±—É—á–∞—é—â–µ–≥–æ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö
image_vectors = np.array([get_image_vector(model, Path(path_to_data) / 'train_images' / img) for img in df_train_filtered['image']])
```

```python
# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–∞–∑–º–µ—Ä–æ–≤ –º–∞—Å—Å–∏–≤–æ–≤
assert text_vectors.shape[0] == image_vectors.shape[0], "–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –≤ text_vectors –∏ image_vectors –¥–æ–ª–∂–Ω–æ —Å–æ–≤–ø–∞–¥–∞—Ç—å"
```

```python
# –ö–æ–Ω–∫–∞—Ç–µ–Ω–∞—Ü–∏—è —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö –∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω—á–µ—Å–∫–∏—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤
combined_vectors = np.hstack((text_vectors.toarray(), image_vectors))
```

```python
# –ü—Ä–∏–≤–µ–¥–µ–Ω–∏–µ –æ—Ü–µ–Ω–∫–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏—è –∫ –¥–∏–∞–ø–∞–∑–æ–Ω—É 0-1
df_expert['expert_score'] = df_expert[['first', 'second', 'third']].mean(axis=1) / 3
```

```python
# –ü–µ—Ä–≤–æ–µ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –≤—Ä–µ–º–µ–Ω–Ω—ã–µ –æ–±—É—á–∞—é—â—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é+—Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
gss = GroupShuffleSplit(n_splits=1, test_size=0.4, random_state=42)
groups = df_train_filtered['image']
train_idx, temp_idx = next(gss.split(combined_vectors, groups=groups))

X_train, X_temp = combined_vectors[train_idx], combined_vectors[temp_idx]
y_train, y_temp = df_expert['expert_score'].iloc[train_idx], df_expert['expert_score'].iloc[temp_idx]

# –í—Ç–æ—Ä–æ–µ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –≤—Ä–µ–º–µ–Ω–Ω–æ–π –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π+—Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
gss = GroupShuffleSplit(n_splits=1, test_size=0.5, random_state=42)
val_idx, test_idx = next(gss.split(X_temp, groups=df_train_filtered['image'].iloc[temp_idx]))

X_val, X_test = X_temp[val_idx], X_temp[test_idx]
y_val, y_test = y_temp.iloc[val_idx], y_temp.iloc[test_idx]

# –ü—Ä–æ–≤–µ—Ä–∏–º —Ä–∞–∑–º–µ—Ä—ã –≤—ã–±–æ—Ä–æ–∫
print(f"–†–∞–∑–º–µ—Ä—ã –≤—ã–±–æ—Ä–æ–∫: –û–±—É—á–∞—é—â–∞—è - {len(X_train)}, –í–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–∞—è - {len(X_val)}, –¢–µ—Å—Ç–æ–≤–∞—è - {len(X_test)}")
```

```python
# –í—ã–≤–æ–¥ —Ä–∞–∑–º–µ—Ä–æ–≤ –≤—ã–±–æ—Ä–æ–∫ –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏
print(f"–†–∞–∑–º–µ—Ä—ã –≤—ã–±–æ—Ä–æ–∫: –û–±—É—á–∞—é—â–∞—è - {len(X_train)}, –í–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–∞—è - {len(X_val)}, –¢–µ—Å—Ç–æ–≤–∞—è - {len(X_test)}")
```

   **–õ–∏–Ω–µ–π–Ω—ã–µ –º–æ–¥–µ–ª–∏ —Å —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–µ–π:**

```python
# Lasso Regression (L1-—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è)
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X_train, y_train)

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
y_val_pred_lasso = lasso_model.predict(X_val)
mae_val_lasso = mean_absolute_error(y_val, y_val_pred_lasso)
print(f"MAE –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –¥–ª—è Lasso —Ä–µ–≥—Ä–µ—Å—Å–∏–∏: {mae_val_lasso}")
```

```python
# –°–æ–∑–¥–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏ –Ω–µ–π—Ä–æ–Ω–Ω–æ–π —Å–µ—Ç–∏
def create_nn_model(input_dim):
    model = Sequential()
    model.add(Dense(64, input_dim=input_dim, activation='relu', kernel_regularizer='l2'))
    model.add(Dropout(0.5))
    model.add(Dense(32, activation='relu', kernel_regularizer='l2'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='linear'))
       
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_absolute_error')
    return model
```

```python
input_dim = X_train.shape[1]
nn_model = create_nn_model(input_dim)
```

```python
# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –Ω–µ–π—Ä–æ–Ω–Ω–æ–π —Å–µ—Ç–∏ –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
nn_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_val, y_val), verbose=1)
```

```python
# –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
y_val_pred_nn = nn_model.predict(X_val).flatten()
```

### 1. –ò–∑–º–µ–Ω–µ–Ω–∏–µ –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä—ã —Å–µ—Ç–∏
–ò–∑–º–µ–Ω–µ–Ω–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ª–æ–µ–≤ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ–π—Ä–æ–Ω–æ–≤ –≤ –∫–∞–∂–¥–æ–º —Å–ª–æ–µ.

```python
from sklearn.model_selection import GroupShuffleSplit, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.base import BaseEstimator, RegressorMixin
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam
```

```python
# –ü—Ä–æ–≤–µ—Ä–∏–º —Ä–∞–∑–º–µ—Ä—ã –≤—ã–±–æ—Ä–æ–∫
print(f"–†–∞–∑–º–µ—Ä—ã –≤—ã–±–æ—Ä–æ–∫: –û–±—É—á–∞—é—â–∞—è - {len(X_train)}, –í–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–∞—è - {len(X_val)}, –¢–µ—Å—Ç–æ–≤–∞—è - {len(X_test)}")
```

```python
# –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –º–æ–¥–µ–ª–∏
def create_model(input_dim, neurons=32, dropout_rate=0.3, learning_rate=0.001):
    model = Sequential()
    model.add(Dense(neurons, input_dim=input_dim, activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1))
    
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='mean_absolute_error', metrics=['mae'])
    return model
```

```python
from sklearn.metrics import make_scorer, mean_absolute_error

# –ì–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è –ø–µ—Ä–µ–±–æ—Ä–∞
param_grid = {
    'neurons': [32, 64, 128],
    'dropout_rate': [0.2, 0.3, 0.4],
    'learning_rate': [0.001, 0.01],
    'batch_size': [16, 32],
    'epochs': [50, 100]
}

best_mae = float('inf')
best_params = {}
best_model = None

# –ü–µ—Ä–µ–±–æ—Ä –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
for neurons in param_grid['neurons']:
    for dropout_rate in param_grid['dropout_rate']:
        for learning_rate in param_grid['learning_rate']:
            for batch_size in param_grid['batch_size']:
                for epochs in param_grid['epochs']:
                    print(f"Training model with neurons={neurons}, dropout_rate={dropout_rate}, learning_rate={learning_rate}, batch_size={batch_size}, epochs={epochs}")
```

```python
# –°–æ–∑–¥–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏
model = create_model(input_dim=X_train_scaled.shape[1], neurons=neurons, dropout_rate=dropout_rate, learning_rate=learning_rate)
```

```python
# –†–∞–Ω–Ω—è—è –æ—Å—Ç–∞–Ω–æ–≤–∫–∞
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
history = model.fit(X_train_scaled, y_train,
                                        validation_split=0.2,
                                        epochs=epochs,
                                        batch_size=batch_size,
                                        callbacks=[early_stopping],
                                        verbose=0)
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞ –≤—Å–µ—Ö –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π –≤ –ø–∞–ø–∫–µ
def get_image_paths_from_folder(folder_path):
    image_paths = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
                image_paths.append(os.path.join(root, file))
    return image_paths

# –ü—Ä–∏–º–µ—Ä —Ñ—É–Ω–∫—Ü–∏–∏ –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏–∏ —Ç–µ–∫—Å—Ç–∞ (–∑–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –≤–∞—à—É —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—é)
def vectorize_text(text, vectorizer):
    # –í–∞—à–∞ –ª–æ–≥–∏–∫–∞ –¥–ª—è –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏–∏ —Ç–µ–∫—Å—Ç–∞
    return vectorizer.transform([text])

# –ü—Ä–∏–º–µ—Ä —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –º–µ—Ç—Ä–∏–∫–∏ –±–ª–∏–∑–æ—Å—Ç–∏ (–∑–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –≤–∞—à—É —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—é)
def predict_similarity(text_vector, image_vectors, model):
    # –í–∞—à–∞ –ª–æ–≥–∏–∫–∞ –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–µ—Ç—Ä–∏–∫–∏ –±–ª–∏–∑–æ—Å—Ç–∏
    similarities = model.predict_proba([text_vector] * len(image_vectors), image_vectors)
    return similarities

# –û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –ª—É—á—à–µ–≥–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
def get_best_image_with_metric(text_description, model, vectorizer, image_df):
    if not text_description:
        return "–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –∑–∞–ø—Ä–æ—Å: —Ç–µ–∫—Å—Ç–æ–≤–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ –Ω–µ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –ø—É—Å—Ç—ã–º.", None
    
    try:
        # –í–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è —Ç–µ–∫—Å—Ç–æ–≤–æ–≥–æ –æ–ø–∏—Å–∞–Ω–∏—è
        text_vector = vectorize_text(text_description, vectorizer)
        
        # –ü–æ–ª—É—á–µ–Ω–∏–µ –∑–∞—Ä–∞–Ω–µ–µ –ø–æ–¥–≥–æ—Ç–æ–≤–ª–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π
        image_vectors = image_df['image_vector'].tolist()
        
        # –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫ –±–ª–∏–∑–æ—Å—Ç–∏
        similarities = predict_similarity(text_vector, image_vectors, model)
        
        # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫ –≤ DataFrame
        image_df['similarity'] = similarities
        
        # –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø–æ –º–µ—Ç—Ä–∏–∫–µ –∏ –≤—ã–±–æ—Ä –ª—É—á—à–µ–≥–æ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è
        best_image_row = image_df.sort_values(by='similarity', ascending=False).iloc[0]
        best_image_path = best_image_row['image_path']
        best_similarity = best_image_row['similarity']
        
        return best_image_path, best_similarity
    
    except Exception as e:
        return f"–ü—Ä–æ–∏–∑–æ—à–ª–∞ –æ—à–∏–±–∫–∞: {e}", None
```
